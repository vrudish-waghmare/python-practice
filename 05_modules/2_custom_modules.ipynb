# Custom Packages and Modules
- `modules`: A python file containing reusable code (e.g. math.py)
- `package`: A directory containing modules and a special `__init__.py` file.
- `sub-packages`: Packages nested within other packages

```python
# Create a package 'math_operations' under which create __init__.py file and two sub folders 'basic' and 'advanced'. under the basic folder create files 'add.py', 'sub.py', '__init__.py and write the code inside the files. do the same thing for advanced sub folder. 
```

```python
from math_operations.basic import add, subtract
print(add(2, 4))
print(subtract(5, 3))
```

```python
from math_operations.advanced import multiply, divide
print(multiply(2, 3))
print(divide(5, 2))
```
