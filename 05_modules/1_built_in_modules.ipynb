# Built-In Modules 
In python, modules and packages help organize and reuse code. python's standard library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

### array

```python
import array
arr = array.array('i', [1, 2, 3, 4])
print(arr)
```

```python
print(type(arr))
```

### math

```python
import math
print(math.sqrt(16))
print(math.pi)
print(math.floor(3.6))
print(math.ceil(4.6))
```

### random

```python
import random 
print(random.randint(1, 10))
print(random.choice(["orange", "pineapple", "banana", "cherry"]))
```

### os

```python
import os   # to work with files and directories
print(os.getcwd())
```

### shutil

```python
# High level operations on files and collections of files
import shutil
shutil.copyfile("source.txt", "destination.txt")
```

### json

```python
import json
data = {
    "name": "Vrudish",
    "age": 24,
    "location":"Nagpur"
}

# convert this data into json
json_string = json.dumps(data)
print(json_string)
print(type(json_string))

# convert json data into python dictionary
parsed_data = json.loads(json_string)
print(parsed_data)
print(type(parsed_data))
```

### csv

```python
import csv

with open("example.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "age"])
    writer.writerow(["Vrudish",24])

with open("example.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

### datetime

```python
from datetime import datetime, timedelta 

now = datetime.now()
print(now)

yesterday = now - timedelta(days=1)
print(yesterday)
```

### time

```python
import time 
print(time.time())
time.sleep(2)
print(time.time())
```

### re

```python
import re

pattern = r'\d+'
text = "There are 123 apples 456"
match = re.search(pattern, text)
print(match.group())
```
