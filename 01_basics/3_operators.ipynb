### Arithmethic Operators
- It is used to perform arithmetic operations like addition, subtraction etc.

```python
a= 10
b = 5

add_result=a+b   # addition
sub_result=a-b   # subtraction 
mult_result=a*b  # multiplication
div_result=a/b   # division
floor_div_result=a//b  # floor division
modulus_result=a%b  # modulus operation

exponent_result=a**b  # Exponentiation


print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

### Logical Operators
- It is used to combine two statements

```python
## and, or, not
X = True 
Y = True 

result = X and Y 
print(result)
```

```python
X = False 
Y = True 
result = X and Y 
print(result)
```

```python
## OR
X=False
Y=False

result = X or Y
print(result)
```

```python
# Not operator
X=False
not X
```

### Identity Operators

```python
# to check if two objects are same with same memory location
# is 
# is not 

x = ["Vrudish", "Rohan"]  # it is a python list to store multiple values
y = ["Vrudish", "Rohan"]
z = x 

print(x is z) 
print(x is y)
print(x == y)
```

Difference between is and == 
* is - checks if both variables points on the same memory 
* == - checks if both variables have same values

### Membership Operators
- It is used to check if a sequence is present in an object or not using `in` and `not in` operators

```python
friend_list = ["vrudish", "rohan", "sohan", "mohan"]
# check if rohan is a friend 
print("rohan" in friend_list)
```

```python
# check if "sonali" is not a friend
print("sonali" not in friend_list)
```
