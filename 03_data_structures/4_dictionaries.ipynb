# Dictionary
Dictionaries are unordered collections of items. They store data in key-value pairs. Key must be unique and immutable (e.g strings, numbers or tuples), while values can be of any type. 

### Creating Dictionary

```python
# empty dict
empty_dict = {}
print(type(empty_dict))
```

```python
student = {"name": "Vrudish", "age": 24, "grade": "A"}
print(student)
```

### Accessing Dictionary elements

```python
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

### Modifying dictionary elements

```python
# update value for the key
student["age"] = 25
```

```python
# add new key-value pairs
student["address"] = "India"
student
```

```python
# delete key and value pairs
del student["address"]
```

```python
student
```

### Dictionary methods

```python
keys = student.keys()
print(keys)
values = student.values()
print(values)
items = student.items()
print(items)
```

### Shallow copy

```python
student_copy = student.copy()
print(student_copy)
```

### Iterating over dictionaries

```python
# Iterating over keys
for key in student.keys():
    print(key) 
```

```python
# Iterating over values
for value in student.values():
    print(value)
```

```python
# Iterating over key-value pairs
for key, value in student.items():
    print(f"{key}: {value}")
```

### Nested Dicitonaries

```python
students = {
    "student1": {"name": "vrudish", "age": 24},
    "student2": {"name": "rohan", "age": 29}
}
print(students)
```

```python
## Access values
print(students["student1"]["name"])
print(students["student1"]["age"])
```

### Iterating over nested dictionaries

```python
for student_id, student_info in students.items():
    print(f"{student_id}:")
    for key, value in student_info.items():
        print(f"{key}: {value}")
    print()
```

### Dictionary Comprehension

```python
squares = {i:i**2 for i in range(1, 11)}
squares
```

```python
evens = {i:i**2 for i in range(1, 11) if i%2==0}
evens
```

```python
# Practical Example
# use a dictionary to count the frequency of elements in list
numbers = [1, 2, 3, 4, 4, 5, 5, 3, 2, 3, 5, 5, 6, 6, 3]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```
