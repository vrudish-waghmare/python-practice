# Sets
Sets are a built-in data type in Python used to store collections of unique items.
They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries and performing mathematical set operatiosn like union, intersection, difference and symmetric difference.

### Create a set

```python
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
# empty set
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1, 2, 3, 4, 5, 6])
print(my_set)
```

### Basics Sets Operations

```python
## Adding and removing elements in set
my_set.add(7)
print(my_set)
```

```python
my_set.remove(3)
my_set
```

```python
my_set.discard(4)
```

```python
my_set
```

```python
# Pop method
removed_ele = my_set.pop()
removed_ele
```

```python
# clear all elements
my_set.clear()
```

```python
my_set
```

```python
## Set membership test
my_set = set([1, 2, 3, 4, 5, 6])
print(3 in my_set)
print(5 in my_set)
```

### Mathematical Operations

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

## union
union_set = set1.union(set2)
print(union_set)

## intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

# difference
difference_set = set1.difference(set2)
print(difference_set)

# symmetric difference
symmetric_set = set1.symmetric_difference(set2)
print(symmetric_set)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}

# issubset
print(set2.issubset(set1))

# issuperset
print(set1.issuperset(set2))
```

```python
### Counting Unique words in text
text = "In this tutorial we are discussing about sets. please follow this steps"
words_list = text.split(" ")
unique_words = set(words_list)
print(unique_words)
print(len(unique_words))
```
