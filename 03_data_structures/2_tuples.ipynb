# Tuples
- Tuples are ordered collection of items that are immutable. They are similar to lists, but their immutabilityb makes them different

### Creating a tuple

```python
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
numbers = tuple([1, 2, 3, 4, 5, 6])
numbers
```

```python
list((1, 2, 3, 4, 5, 6))
```

```python
# mixed datatypes
mixed_tuple = (1, "Hello", 3.14, True)
mixed_tuple
```

### Accessing Tuple elements

```python
print(mixed_tuple[0])
print(mixed_tuple[-1])
print(mixed_tuple[1:])
print(mixed_tuple[::-1])
```

### Tuple Operations

```python
tpl1 = (1, 2, 3, 4, 5)
tpl1 * 3
```

```python
tpl1 + ("a", "b", "c")
```

### Immutable Nature of Tuples
- Tuples are immutable, meaning their elements cannot be changed once assigned.

```python
tpl1[0] = 100
```

### Tuple Methods

```python
## count()
print(tpl1.count(1))
```

```python
## index()
print(tpl1.index(3))
```

### Packing and Unpacking tuple

```python
## Packing
packed_tuple = "Hello", 3.14, 3
packed_tuple
```

```python
## Unpacking 
a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers 
print(first)
print(middle)
print(last)
```

### Nested Tuple 

```python
tpl = ((1, 2), (3, 4), (5, 6))
tpl
```

```python
## Accessing nested tuple elements
tpl[0][1]
```
