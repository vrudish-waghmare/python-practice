# File Operations- Read and Write Files
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary.

```python
#Read a whole file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Writing a file (Overwriting)
with open("example.txt", "w") as file:
    file.write("Hello World!")
    file.write("This is a new line.")
```

```python
# Writing a file (without overwriting)
with open("example.txt", "a") as file:
    file.write("Append operation taking place!")
```

```python
# Writing a list of lines to a file
lines = ['First line\n', 'Second line \n', 'Thrid line \n']
with open("example.txt", "a") as file:
    file.writelines(lines)
```

```python
# Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
# Let's create a copy file function to copy content from source to destination
def copyFile(sourcePath, destinationPath):
    """
    This function used to copy the content of source file to a destination file
    """
    # get the content from the source file
    with open(sourcePath, "r") as source_file:
        content = source_file.read()
        
    with open(destinationPath, "w") as destination_file:
        destination_file.write(content)
```

```python
# calling the function
copyFile("example.txt", "destination.txt")
```

```python
# Read a text file and count the number of lines, words and characters. 
def count_text_file(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count
```

```python
# calling the function
file_path = "example.txt"
lines, words, characters = count_text_file(file_path)
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```

`w+` mode in python is used to open a file for both reading and writing. If the file does not exist, it will be created. if the file exists, its content is truncated (i.e the file is overwritten)

```python
# Writing and then reading a file
with open("example.txt", "w+") as file:
    file.write("Hello World\n")
    file.write("This is a new line \n")

    # Move the file cursor to the beginning
    file.seek(0)

    # Read the content of the file
    content = file.read()
    print(content)
```
