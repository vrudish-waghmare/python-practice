# Functions
A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readibility.

```python
# syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
def check_even_or_odd(num):
    """This function checks if a number is even or odd"""
    if num%2==0:
        print("This number is even")
    else:
        print("This number is odd")
```

```python
# call this function
check_even_or_odd(4)
```

```python
# function with multiple parameters
def add(num1, num2):
    """This function adds two numbers and return their sum"""
    return num1+num2 

result = add(4, 5)
print(result)
```

```python
# Default Parameters
def greet(name="Guest"):
    """This function greets user by taking name as a parameter"""
    print(f"Hello {name}, Welcome To the paradise")
greet("Vrudish")
```

```python
greet()
```

```python
### Variable Length Arguments 
# Positional and Keywords Arguments

def print_numbers(*args):
    for i in args:
        print(i)
```

```python
print_numbers(1, 2, 3, 4, 5, 6, "Vrudish")
```

```python
## Keyword arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Vrudish", age=24, location="Nagpur")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional argument: {val}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
    
```

```python
print_details(1, 2, 3, 4, name="Vrudish", age = "Rohan")
```

```python
# Return statement
def multiply(a, b):
    return a*b
```

```python
multiply(2, 3)
```

```python
# Return multiple parameters
def arithmetic_operation(a, b):
    return a+b, a-b, a*b, a/b
```

```python
arithmetic_operation(2, 3)
```

```python
# Type Hints
def multiply(num1:int, num2:int) -> int:
    return num1 * num2
```

```python
# Type Hints for List and Dict
from typing import List, Dict 
def display_names(names:List[str]) -> None:
    for name in names:
        print(name, '\n')
```

```python
display_names(["Vrudish", "Rohan", "Sohan"])
```
