# Map Function
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object.(iterator). This is particularly useful for transforming data in a lsit comprehensivly.

```python
def square(x):
    return x*x 

square(2)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
sqauared_numbers = list(map(square, numbers))
sqauared_numbers
```

```python
# lambda function with map
list(map(lambda n:n**2, numbers))
```

```python
# Map multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
added_numbers = list(map(lambda a, b: a+b, numbers1, numbers2))
added_numbers
```

```python
# map() to convert list of strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))
int_numbers
```

```python
words = ["apple", "banana", "cherry"]
upper_words = list(map(str.upper, words))
upper_words
```

```python
people = [
    {'name': "Vrudsih", 'age': 24},
    {"name": "Rohan", "age": 29}
]

def get_name(people):
    return people['name']

list(map(get_name, people))
```
