# Lambda
Lambda functions are small anonymous functons defined using the **lambda** keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to hight-order functions.

```python
# Syntax
lambda arguments: expression
```

```python
def addition(a, b):
    return a+b
```

```python
# A better way to write single statement function is using lambda function
add = lambda a, b: a+b
add(2, 3)
```

```python
type(add)
```

```python
# more that 2 arguments
multiply = lambda a, b, c : a * b * c 
multiply(2, 3, 4)
```

```python
# using lambda in map function 
# map() - applies a function to all items in a list
numbers = [1, 2, 3, 4, 5]
numbers_square = list(map(lambda num: num**2, numbers))
print(numbers_square)
```
