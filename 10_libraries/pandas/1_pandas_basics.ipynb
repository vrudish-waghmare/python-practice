# Pandas 
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
print(pd.__version__)
```

### Series

```python
# Series - A pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd 
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
series[0]
```

```python
# Create a series from dictionary
data = {"a": 1, "b": 2, "c": 3}
series_dict = pd.Series(data)
print(series_dict)
```

### DataFrame

```python
# Create a dataframe from a dictionary of list
data = {
    "Name": ["Vrudish", "Rohan", "Sohan"],
    "Age": [24, 25, 26],
    "City": ["Nagpur", "Pune", "Kolkata"]
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Create a dataframe from a list of dictionaries
data = [
    {"Name": "Vrudish", "Age": 24, "City": "Nagpur"},
    {"Name": "Rohan", "Age": 25, "City": "Pune"},
    {"Name": "Sohan", "Age": 26, "City": "Kolkata"}
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
