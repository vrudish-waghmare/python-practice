# Data Manipulation and Analysis with Pandas
Data manipulation and analysis are key tasks in any data science or data analysis project. Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean, transform, and extract insights from data.

```python
import pandas as pd
print(pd.__version__)
```

### Handling Missing Values

```python
# check missing values
df.isnull().sum()
```

```python
# filling missing values with 0
df_filled = df.fillna(0)
```

```python
df_filled.isnull().sum()
```

```python
# filling missing values with the mean of the column
df["Sales_fillNA"] = df["Sales"].fillna(df["Sales"].mean())
df.head()
```

### Renaming columns

```python
df = df.rename(columns={"Date": "Sales Date"})
df.head()
```

### Change Datatypes


```python
df.dtypes
```

```python
# change datatypes
df["Value_new"] = df["Value"].fillna(df["Value"].mean()).astype(int)
```

```python
df.head()
```

```python
# transforming column
df["New_Value"] = df["Value"].apply(lambda x: x*2)
df.head()
```

### Data Aggregating and Grouping

```python
df.head()
```

```python
# calculate the average sales per Product
grouped_mean = df.groupby("Product")["Value"].mean()
print(grouped_mean)
```

```python
# Calculate the total sales per product and region
grouped_sum = df.groupby(["Product", "Region"])["Value"].sum()
print(grouped_sum)
```

```python
# Calculate the average sales per product and region
df.groupby(["Product", "Region"])["Value"].mean()
```

```python
## aggregate multiple functions
groudped_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
groudped_agg
```

### Merging and Joining DataFrames

```python
# Create sample DataFrames
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
# Merge DataFrame on the 'Key columns'
pd.merge(df1, df2, on="Key", how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
