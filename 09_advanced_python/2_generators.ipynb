# Generators 
Generators are a simple way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memeory.

```python
def nth_square(n):
    for i in range(1, n+1):
        yield i**2
```

```python
nth_square(5)
```

```python
a = nth_square(5)
```

```python
next(a)
```

```python
for square in nth_square(5):
    print(square)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

### Practical example: Reading Large Files
Generators are particularly useful for reading large files because they allow you to process one line at a time without loading the entire file into memory.

```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line 
```

```python
file_path = 'large_file.txt'
for line in read_large_file(file_path):
    print(line.strip())
```
