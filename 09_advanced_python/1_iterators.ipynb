# Iterators
Iterators are advanced python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1, 2, 3, 4, 5, 6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
# Iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
# Iterate through all the elements
next(iterator)
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("Finished")
```

```python
# String Iterator
my_name = "Vrudish" 
name_iter = iter(my_name)

print(next(name_iter))  # Output: V
print(next(name_iter))  # Output: r
```
