# Decorators 
Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. 


### Function Copy 

```python
def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python
wel = welcome   # function copy
print(wel())
del welcome 
print(wel())
```

### Closures
A closure in Python is a nested function that allows access to variables from its enclosing scope even after the outer function has finished execution.

```python
def main_welcome(msg):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome("Hello, Closure")
```

```python
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func("Welcome everyone to this tutorial")
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
def main_welcome(func, lst):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(func(lst))
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(len, [1, 2, 3, 4, 5])
```

### Decorator

```python
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("This is an advanced python course")

main_welcome(course_introduction)
```

```python
@main_welcome
def course_introduction():
    print("This is an advanced python course")
```

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators With arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello!")
```

```python
say_hello()
```
